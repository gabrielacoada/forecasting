## Question 1: Forecasting Scenario Analysis

> *For each of the following scenarios, discuss the decision environment, the nature of the object to be forecast, the forecast type, the forecast horizon, the loss function, the information set, and what sorts of simple or complex forecasting approaches you might entertain.*

### (a) Airborne Analytics — Fighter Plane Contract

**Decision Environment:** You work for a specialized mutual fund investing exclusively in airline stocks. A newly rich oil-producing country has requested bids on a contract to deliver 30 fighter planes, and only two companies submitted bids. The stock of the successful bidder will likely rise, and you must decide how to position the fund's portfolio. The stakes are high — a correct call could yield substantial returns, while a wrong call leads to opportunity cost or losses. This is a one-shot, event-driven decision with a binary outcome.

**Forecast Object:** The object to be forecast is the probability that each of the two bidding companies wins the contract. Equivalently, you are forecasting a discrete event (which company wins), not a continuous time series. This is a cross-sectional, qualitative prediction rather than a standard time series forecasting problem.

**Forecast Type:** Event probability forecast. You need to assess $P(\text{Company A wins})$ vs. $P(\text{Company B wins})$. This naturally leads to a point estimate of the probability (or a binary prediction of the winner), along with a subjective assessment of confidence.

**Forecast Horizon:** Short to medium term — the bid decision will be announced at some specific future date (perhaps weeks or months). This is not a recurring forecast; it is a one-time event.

**Loss Function:** Asymmetric. If you correctly predict the winner and invest accordingly, the fund benefits from the stock price increase. If you predict incorrectly, you may have shifted portfolio weights toward the loser (whose stock might decline) or simply missed the gain on the winner. The loss function is likely closer to **linlin (asymmetric)** — the cost of missing the winning stock is the foregone gain, while overweighting the losing stock incurs actual losses. Under this asymmetric setup, as Professor Pesavento emphasized with the stock investing example, "you're not equally happy if you under-predict your stock or over-predict your stock."

**Information Set:** This is unusual for a forecasting class — the relevant information is not a historical time series but rather qualitative intelligence: the track records of the two companies, their political connections with the oil-producing country, past defense contract outcomes, pricing history, lobbying efforts, geopolitical relationships, and possibly insider knowledge about the bid terms. You might also use financial market data (e.g., movements in the stocks of the two bidders around the announcement of the bid, which could reflect market expectations).

**Forecasting Approaches:**
- **Simple:** Assign equal probability (50/50) and hedge by investing in both companies. This is the "naive" forecast — analogous to forecasting the unconditional mean.
- **Moderate:** Gather qualitative intelligence (defense industry analysis, geopolitical analysis, past contract outcomes) and form a subjective probability assessment. Use financial market signals — if one company's stock has already moved more on the news of the bid, the market may be pricing in a higher probability of that company winning.
- **Complex:** Build a logit/probit model of defense contract outcomes using historical data on similar contracts (company size, political ties, price competitiveness). However, with only two bidders and a unique contract, there may not be enough comparable historical data for a formal statistical model.

### (b) Office of Management and Budget — Tax Revenue Forecast

**Decision Environment:** You work for the OMB and must forecast tax revenues for the upcoming fiscal year. The institutional context is critical: the president wants high forecasts to maintain funding for his social programs, but not so high that a large deficit materializes and damages his reelection prospects. Additionally, Congress will scrutinize your methodology. The decision environment is therefore politically charged — you face pressure from multiple directions and your forecast has direct policy consequences.

**Forecast Object:** Total U.S. federal tax revenues for the upcoming fiscal year. This is a flow variable (accumulated over the fiscal year), a continuous, aggregate macroeconomic quantity. It comprises income taxes, corporate taxes, payroll taxes, excise taxes, etc.

**Forecast Type:** Point forecast, likely accompanied by an interval or range estimate. In practice, the OMB produces baseline revenue projections under current-law assumptions, with scenarios for different economic outcomes. Given the political stakes, a density forecast (showing the full range of possibilities) might be more defensible to Congress than a single point estimate.

**Forecast Horizon:** One year ahead — the upcoming fiscal year. This is a medium-term forecast. The horizon is fixed by the budget calendar.

**Loss Function:** This is the most interesting aspect of this scenario. The loss function is highly **asymmetric and politically complex**:
- **Overprediction (forecast too high):** The president spends based on optimistic revenue projections, runs a deficit, and is seen as fiscally irresponsible → political damage, loss of reelection support.
- **Underprediction (forecast too low):** Social programs lose funding because the budget doesn't allocate enough → the president's policy agenda suffers.
- **Methodological scrutiny:** If Congress finds fault with your procedures, the budget itself can be undermined.

The loss is therefore **asymmetric linlin** — but the direction of asymmetry is ambiguous. The president prefers a slight overestimate (to fund programs), but the cost of a *large* overestimate is severe (deficit, political fallout). This suggests a loss function that penalizes large positive errors (overforecasting revenue) much more heavily than moderate positive errors, while also penalizing underprediction. The optimal forecast should lean slightly above the conditional mean, but not excessively. As Pesavento taught, under linlin loss the optimal forecast is a quantile — here, one slightly above the median.

Critically, the forecast also needs to be **defensible** — any bias introduced to serve political goals must be subtle enough to withstand Congressional scrutiny. This is a constraint that most forecasting textbooks don't address: the forecast must be not only accurate but also perceived as unbiased.

**Information Set:** Very rich. The OMB has access to:
- Historical tax revenue data (monthly, quarterly, annual) — a long time series
- Macroeconomic forecasts of GDP growth, unemployment, inflation, interest rates
- Tax code details and any proposed changes
- Corporate earnings forecasts, labor market data
- CBO (Congressional Budget Office) independent forecasts for comparison
- Survey of Professional Forecasters (SPF) for macro assumptions

**Forecasting Approaches:**
- **Simple:** Extrapolate the historical trend in tax revenues using a deterministic trend model ($y_t = c + \beta t + u_t$), possibly with seasonal adjustment for monthly/quarterly data.
- **Moderate:** Build an ARMA model of tax revenue growth, conditioning on macroeconomic variables (GDP growth, inflation, employment). Use the pseudo out-of-sample framework to validate.
- **Complex:** A structural macroeconomic model that links tax revenues to GDP, employment, capital gains, corporate profits, and the tax code. The OMB likely uses something close to this — a large-scale macro model with a tax module. This also allows scenario analysis ("what if GDP growth is 1% lower?").

### (c) D&D Advertising — Predicting Teen Trends

**Decision Environment:** You work for an advertising firm and must create an ad targeted at teenagers. You need to forecast what teenagers currently consider "cool" and incorporate that into the ad. Success brings glory and future clients; failure means lost revenue and potentially losing the client. This is a high-stakes, winner-take-all situation.

**Forecast Object:** The object to be forecast is qualitative and latent — teenage cultural preferences, trends, and tastes. Unlike tax revenues or stock prices, this is not a well-defined numerical quantity. It is more of a sentiment or cultural state that must be inferred from indirect signals. This is perhaps the hardest type of object to forecast because it is not directly measurable and changes rapidly.

**Forecast Type:** Qualitative assessment / scenario-based. You are not producing a number — you are producing a description of the cultural zeitgeist among teenagers. This is closer to a scenario forecast or a subjective probability distribution over possible "cool" themes.

**Forecast Horizon:** Very short term — you need to know what is cool *right now* or in the very near future (when the ad airs). But there is a production lag: by the time the ad is produced and aired, "cool" may have shifted. So the effective horizon is the production lag (perhaps weeks to months). In rapidly evolving teen culture, even a few weeks can be a long time.

**Loss Function:** Strongly asymmetric and nonlinear. A "hit" — correctly identifying the trend — delivers enormous returns (viral success, new clients). A "miss" delivers substantial losses (client dissatisfaction, lost account). This is closer to a **binary outcome with asymmetric payoffs**: the upside of a correct prediction may exceed the downside of a wrong one (you can get many future clients from one success), but a miss can lose the current client entirely. The loss function resembles **all-or-nothing** rather than the smooth quadratic or linlin forms — small forecast errors may be just as costly as large ones if the ad simply doesn't resonate.

**Information Set:** Unconventional compared to time series data:
- Social media trends (TikTok, Instagram, YouTube — trending content, hashtags, viral videos)
- Focus groups with actual teenagers
- Market research data on teen spending and brand preferences
- Music charts, fashion trends, entertainment trends
- Historical advertising success/failure data for teen-targeted products
- Cultural commentators and trend forecasters

This is largely cross-sectional/qualitative information, not a neat time series. However, some of these signals (social media engagement metrics, search trends) can be tracked over time.

**Forecasting Approaches:**
- **Simple:** Look at what was "cool" recently and assume it persists — essentially a random walk or naive forecast of cultural trends.
- **Moderate:** Conduct focus groups, survey teenagers, analyze social media data systematically. Use trend analysis on Google Trends or TikTok engagement data to identify what is currently rising in popularity.
- **Complex:** Build a model of cultural trend diffusion (e.g., how trends spread from early adopters to the mainstream), combining social media analytics with demographic data. Use NLP on social media text to quantify sentiment around different cultural themes. However, the professor emphasized that model complexity must serve a purpose — the Southern Company example shows that even sophisticated firms often use simple, explainable models.

## Question 2: Trend Model — 500 Daily vs. 50 Annual Observations

> *Suppose it's the last day of the year. You are using a trend model to produce a 1-year-ahead (end-of-year) forecast of a stock (as opposed to flow) variable observed daily. Would you prefer to estimate your forecasting model using the most recent 500 daily observations (and then forecast 365 steps ahead) or 50 annual end-of-year observations (and then forecast 1 step ahead)?*

### Answer

**I would prefer to use the 50 annual end-of-year observations and forecast 1 step ahead**, despite 500 being a much larger sample size than 50. The reasoning involves several interrelated considerations:

#### 1. Forecast Horizon Matters Enormously

With the 500 daily observations approach, the trend model $y_t = c + \beta t + u_t$ is estimated at the daily frequency, and then you must project the trend **365 steps ahead** to reach the next end-of-year value. As we learned in class, forecast errors accumulate with the forecast horizon $h$. Even small estimation errors in $\hat{\beta}$ are multiplied by 365, and the forecast error variance grows rapidly. The confidence interval around a 365-step-ahead forecast will be enormously wide.

With the 50 annual observations approach, you estimate the trend at the annual frequency and project only **1 step ahead**. The forecast error is $e_{T+1,T} = u_{T+1}$ — just one period of noise. The confidence interval is much tighter.

#### 2. Stock Variable vs. Flow Variable — The Key Distinction

The question specifies a **stock** variable — measured at a point in time (end-of-year), not accumulated over a period. This is the critical insight. A stock variable at the end of the year is a single snapshot. All the intra-year daily fluctuations are irrelevant to the end-of-year value — they are noise from the perspective of forecasting the annual end-of-year level.

If this were a **flow** variable (e.g., total annual GDP), the daily observations would contain useful information because each daily observation contributes to the annual total. For a stock variable, the 499 intra-year daily observations are essentially noise that does not help predict where the variable will be at the end of the *next* year.

#### 3. Larger Sample Size $\neq$ Better Forecast

It might seem that $n = 500$ is far superior to $n = 50$ because standard errors shrink as $1/\sqrt{n}$. However, this logic breaks down for several reasons:

- **Serial correlation:** Daily observations of a stock variable are highly serially correlated. As the professor emphasized, time series data violates the i.i.d. assumption — observations close in time are dependent. With extreme positive serial correlation (DW near 0), the effective sample size is much less than 500. Each daily observation adds very little independent information beyond what nearby days already provide.

- **Relevant variation:** For forecasting the end-of-year level, what matters is the year-to-year variation in the trend. The 50 annual observations capture 50 independent (or nearly independent) annual data points. The 500 daily observations, despite being nominally more numerous, capture essentially the same year-to-year variation but contaminated with high-frequency noise.

- **Forecast horizon penalty:** The gain from a slightly more precise $\hat{\beta}$ estimate is overwhelmed by the cost of forecasting 365 steps ahead. Even if $\hat{\beta}$ is estimated more precisely with daily data, the long forecast horizon amplifies all sources of uncertainty.

#### 4. Practical Considerations

- A trend model at daily frequency will have residuals with extremely low Durbin-Watson statistics (near zero), indicating severe serial correlation. This violates the model assumptions and makes inference unreliable — exactly the issue the professor demonstrated with the GDP trend regression.
- Daily data may also exhibit seasonality (day-of-week effects, month-of-year patterns) that the simple trend model ignores, introducing systematic forecast error.
- The 50 annual observations provide a cleaner, more appropriate sample for estimating the annual trend.

#### Summary

The 50 annual observations are preferred because: (1) the forecast horizon is only 1 step ahead rather than 365; (2) the variable is a stock, so intra-year observations are noise for end-of-year forecasting; (3) serial correlation in daily data means the effective sample size is far less than 500; and (4) the forecast error variance grows with the horizon, overwhelming any precision gains from the larger nominal sample.

## Question 3: Seasonal Adjustment

> *One way to deal with seasonality in a series is simply to remove it, and then to model and forecast the seasonally adjusted series.*

### (a) Why You Might Use Seasonally Adjusted Data

One important reason to use seasonally adjusted data is when your interest centers on **identifying and forecasting nonseasonal fluctuations**, particularly business cycle dynamics. In macroeconomics, policymakers and analysts are often primarily interested in whether the economy is expanding or contracting — i.e., the cyclical component of a series — rather than in predictable seasonal patterns.

For example, GDP typically rises in Q4 due to holiday spending and falls in Q1. These seasonal swings are large and predictable, but they tell us nothing about whether the economy is entering a recession. If we work with the raw (not seasonally adjusted) data, the large seasonal variation can mask the underlying business cycle movements. Seasonal adjustment removes this predictable calendar-driven variation, allowing us to focus on quarter-to-quarter changes that reflect genuine economic shifts.

As covered in class, the professor noted that there are two choices when dealing with seasonality: use seasonally adjusted data (remove seasonality first) or model seasonality explicitly with dummy variables and forecast all variation. The first approach is appropriate when seasonality is not the object of interest — as is often the case in macroeconomic policy analysis. Agencies like the BLS and the Federal Reserve report seasonally adjusted figures precisely because month-to-month or quarter-to-quarter comparisons are more informative for policy decisions when the seasonal component has been stripped away.

A second, more technical reason is that seasonal adjustment can simplify the modeling task. Rather than including $s - 1$ seasonal dummy variables (or seasonal ARIMA terms) in every model, one can work with the SA series and focus on modeling the trend-cycle dynamics. This is especially useful when combining multiple series, as different series may have different seasonal patterns.

### (b) U.S. Census Bureau Seasonal Adjustment Procedure

The latest U.S. Census Bureau seasonal adjustment procedure is **X-13ARIMA-SEATS**, first released in 2012 as the successor to X-12-ARIMA. The most recent version is Version 1.1, Build 62 (released July 2025). The name reflects the integration of two distinct traditions: the Census Bureau's filter-based X-11 approach and the Bank of Spain's model-based SEATS (Signal Extraction in ARIMA Time Series) approach.

#### How It Works

X-13ARIMA-SEATS operates in two stages:

**Stage 1 — RegARIMA Pre-adjustment:** A regression model with ARIMA errors (regARIMA) is fitted to the raw series. This step handles:
- **Calendar effects:** Trading day variations (different months have different day-of-week compositions) and moving holidays (Easter, Labor Day) are modeled with regression variables.
- **Outlier detection:** The program automatically identifies additive outliers (AO), level shifts (LS), and transitory changes (TC) — analogous to the structural breaks we studied in class with the Chow test.
- **Forecasts and backcasts:** The ARIMA model extends the series forward and backward to stabilize the moving-average filters at the endpoints.

The output is a "linearized" series — the original data with deterministic and outlier effects removed.

**Stage 2 — Seasonal Decomposition:** The user chooses between two decomposition methods:
- **X-11 (filter-based):** Uses iterative weighted symmetric moving averages (Henderson trend filters and seasonal moving averages) to decompose the series into trend-cycle, seasonal, and irregular components. This is a nonparametric approach.
- **SEATS (model-based):** Derives Wiener-Kolmogorov optimal signal extraction filters from the ARIMA model to decompose the series into unobserved components. This is a parametric approach where the decomposition is implied by the model.

The final seasonally adjusted series is produced by subtracting (additive) or dividing (multiplicative) the seasonal and calendar factors from the original data.

#### Historical Lineage

The procedure has evolved over more than 60 years:
- **X-11** (1965, Census Bureau): The original iterative moving-average decomposition method.
- **X-11-ARIMA** (1975, Statistics Canada): Added ARIMA-based forecasts to extend the series, reducing endpoint instability.
- **X-12-ARIMA** (1998, Census Bureau): Added full regARIMA modeling for calendar effects and outlier detection.
- **TRAMO-SEATS** (1990s, Bank of Spain): A fully model-based alternative.
- **X-13ARIMA-SEATS** (2012, Census Bureau): Unified both approaches into a single program.

#### Key Features

- **Automatic model selection:** The program can automatically choose the ARIMA order and identify significant calendar/outlier effects.
- **Diagnostics:** Spectrum-based tests for residual seasonality, sliding spans for revision stability, and standard goodness-of-fit measures.
- **Flexibility:** Supports additive, multiplicative, log-additive, and pseudo-additive decompositions.

X-13ARIMA-SEATS remains the standard tool used by U.S. statistical agencies (Bureau of Labor Statistics, Federal Reserve) and is actively maintained. The European alternative, JDemetra+, implements the same algorithms in an open-source Java environment.

**Sources:** U.S. Census Bureau, X-13ARIMA-SEATS Seasonal Adjustment Program (census.gov/data/software/x13as.html); X-13ARIMA-SEATS Reference Manual, Version 1.1.
